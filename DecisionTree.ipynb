```python
df['Embarked'].fillna('S',inplace=True)
```

```python
df.isna().sum()
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
age_mean=df['Age'].mean();
age_std=df['Age'].std()

ramdom_values=np.random.normal(age_mean,age_std,177)
ramdom_values
```

```python
df.loc[df['Age'].isna(),"Age"]=ramdom_values
```

```python
df.isna().sum()
```

```python
df['FamilySize']=df['Parch']+df['SibSp']+1
```

```python
df.head()
```

```python
df['IsAlone']=df['FamilySize']==1
```

```python
df.head()
```

```python
df1=df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df1.drop(columns=['Survived'])
y=df1['Survived'];
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# Label encoding of categorical features
X['Sex']=le.fit_transform(X['Sex'])
X['Embarked']=le.fit_transform(X['Embarked'])
X['IsAlone']=le.fit_transform(X['IsAlone'])
X
```

```python
sns.pairplot(X)
```

```python
X
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(criterion='entropy')
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True)
plt.show()
```

```python

plt.figure(figsize=(15,10))
tree.plot_tree(dt,filled=True,max_depth=3,feature_names=X.columns)
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score

# Importing classification report and confusion matrix from sklearn metrics

# Making predictions

# Printing classification report
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report

sns.heatmap(confusion_matrix(y_pred,y_test),cmap="winter")
confusion_matrix(y_pred,y_test)
```
